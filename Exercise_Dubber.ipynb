```python
#Importing required library
from random import randint
import random
#Creating an empty list
number = []
#Writing a function to generate a random number
def generate_number():
    for i in range(7):
      number.insert(i,randint(1,49))
    a_set = set(number)
    if len(a_set)==7: 
        a_set.clear()
    else:
        number.clear()
        generate_number()
    #Sorting the generated numbers
    number.sort()
    #Assigning bonus number to a variable
    bonus=number[6]
    print(number)
    for i in range(6):
     print(i)
     if number[i] >= 1 and number[i] <= 9:
        bc="Grey"      
        print_nums(i,bc)
     elif number[i] >=10 and number[i] <= 19:
        bc="Blue"
        print_nums(i,bc)
     elif number[i] >=20 and number[i] <= 29:
        bc="Pink"
        print_nums(i,bc)
     elif number[i] >=30 and number[i] <= 39:
        bc="Green"
        print_nums(i,bc)
     else:
        bc="Yellow" 
        print_nums(i,bc)    
    b1.config(state="disabled")
    b2.config(state="active")
    
#Creating a function to display bonus number
def get_bonus():
    l2.config(text=number[6], bg="Cyan")
          
#Creating a function to display first six number
def print_nums(i,color):
      if i==0:  
         p1.config(bg=color, text=number[0])
      elif i==1 :
         p2.config(bg=color, text=number[1]) 
      elif i==2 :
         p3.config(bg=color, text=number[2])  
      elif i==3 :
         p4.config(bg=color, text=number[3]) 
      elif i==4 :
         p5.config(bg=color, text=number[4]) 
      else:
         p6.config(bg=color, text=number[5]) 

#Importing tkinter library
import tkinter as tk
window = tk.Tk()
window.geometry("500x300")
window.config(bg="White")
window.resizable(width=False,height=False)
#Displaying title 
window.title('Welcome to Lottery draw')
 
 
l1 = tk.Label(text="Let's find the lucky ones",font=("Arial",20),bg="Blue",fg="White")
#Displying user interface button to get lottery numbers
b1 = tk.Button(text="Click here to draw Lottery numbers",font=("Arial",15),bg="#A3E4D7",command=generate_number)

#Displying user interface button to get bonus number
b2=  tk.Button(text="Click here to find the bonus draw !!!",font=("Arial",15),bg="#A3E4D7",fg="Black",command=get_bonus, state="disabled")
l2=  tk.Label(text="",font=("Arial",20),bg="White",fg="Black")
p1 = tk.Label(font=("Arial",30),text="")
p2 = tk.Label(font=("Arial",30),text="")
p3 = tk.Label(font=("Arial",30),text="")
p4 = tk.Label(font=("Arial",30),text="")
p5 = tk.Label(font=("Arial",30),text="")
p6 = tk.Label(font=("Arial",30),text="")




 
l1.place(x=140,y=20)
l2.place(x=234, y=235)
b1.place(x=110,y=70)
b2.place(x=120,y=190)
p1.place(x=135,y=130)
p2.place(x=190,y=130)
p3.place(x=245,y=130)
p4.place(x=300,y=130)
p5.place(x=355,y=130)
p6.place(x=410,y=130)


window.mainloop()
```
